```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
# Using Gradient Boosting for advanced tabular modeling
from sklearn.ensemble import GradientBoostingRegressor

# Set seed for academic reproducibility
np.random.seed(101)
n_samples = 1500

# 1. BIOLOGICAL: DNA Polygenic Score (Normalized scale from 0 to 1)
# Represents cumulative genetic variants correlated with neuroplasticity
dna_polygenic_score = np.random.beta(a=5, b=5, size=n_samples)

# 2. BIOLOGICAL: Blood Pressure (Systolic / Diastolic)
# High blood pressure (hypertension) can negatively impact cognitive load scores over time
systolic_bp = np.random.normal(120, 15, n_samples)
diastolic_bp = np.random.normal(80, 10, n_samples)

# 3. COGNITIVE: Text Recall Memory Score (Percentage 0% to 100%)
# Simulating a short-term memory test of recalling text fragments
text_recall_score = np.random.normal(70, 12, n_samples)
text_recall_score = np.clip(text_recall_score, 0, 100)

# Target Variable: Simulating an IQ score influenced by these multimodal features
# Higher DNA score & Memory = positive correlation. Extreme high blood pressure = negative correlation.
noise = np.random.normal(0, 4, n_samples)
simulated_iq = (
    60
    + (35 * dna_polygenic_score)
    + (0.4 * text_recall_score)
    - (0.15 * (systolic_bp - 120))
    + noise
)
simulated_iq = np.clip(simulated_iq, 55, 145)

# Assemble into Master DataFrame
biopsych_df = pd.DataFrame({
    'DNA_Polygenic_Score': dna_polygenic_score,
    'Systolic_BP': systolic_bp,
    'Diastolic_BP': diastolic_bp,
    'Text_Recall_Score': text_recall_score,
    'IQ': simulated_iq
})

print("Multimodal Biopsychological Dataset Created!")
print(biopsych_df.head())
```

```python
# Feature and Target Split
X = biopsych_df.drop(columns=['IQ'])
y = biopsych_df['IQ']

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)

# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and Train Gradient Boosting Model
gb_model = GradientBoostingRegressor(n_estimators=150, learning_rate=0.05, max_depth=4, random_state=101)
gb_model.fit(X_train_scaled, y_train)

# Predict and Evaluate
y_pred = gb_model.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"--- Gradient Boosting Results ---")
print(f"Mean Absolute Error: {mae:.2f} IQ points")
print(f"R² Score: {r2:.2f}")
```

```python
# Extract feature importance
importances = gb_model.feature_importances_
feature_names = X.columns

# Plotting
plt.figure(figsize=(9, 5))
sns.barplot(x=importances, y=feature_names, palette='viridis', hue=feature_names, legend=False)
plt.title('AI Feature Importance: Biological vs. Cognitive Predictors of IQ')
plt.xlabel('Importance Value')
plt.ylabel('Features')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
